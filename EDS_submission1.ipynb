```python
import pandas as pd

%matplotlib inline
```

## Dataset loading for visualization

```python
df = pd.read_csv("C:\\Users\\ashut\\Desktop\\n\\covid-data(EDS).csv")
df.head()
```

```python
df_location = df['location'].unique()
df_location
```

```python
df_Ger = df[df['location'] == 'Germany']
df_Ind = df[df['location'] == 'India']
df_Spa = df[df['location'] == 'Spain']
df_loc = [df_Ger, df_Ind, df_Spa]
df_location = pd.concat(df_loc)
df_location
```

```python
df_location = df_location.reset_index(drop=True)
df_location
```

```python
df_location_na = df_location.isna().sum()
df_location_na[df_location_na.values > 0].sort_values(ascending = False)
```

## Line graph for the dataset

```python
df_location
```

```python
import plotly.express as px

fig = px.line(df_location,  x = "date" ,y = "total_cases", title = 'COVID Cases')
fig.show()
```

```python
fig = px.line(df_location,  x = "date" ,y = "covid_rate", color = 'location', title = 'COVID Cases')
fig.show()
```
